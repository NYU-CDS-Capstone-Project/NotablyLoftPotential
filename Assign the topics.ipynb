## Assign topics to each tweet using the best iteration of each model

```python
import pickle
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

%matplotlib inline

pd.set_option('display.max_info_columns',200)
pd.set_option('display.max_colwidth',200)
pd.set_option('display.max_rows', 100)
plt.style.use('ggplot')
```

### Load the models

```python
climate_model = pickle.load(open("pickles/.p", "rb"))
guns_model = pickle.load(open("pickles/.p", "rb"))
marriage_model = pickle.load(open("pickles/.p", "rb"))
overall_model = pickle.load(open("pickles/.p", "rb"))
```

### Load the dictionaries

```python
climate_dict = pickle.load(open("pickles/climate_dictionary.p", "rb"))
guns_dict = pickle.load(open("pickles/guns_dictionary.p", "rb"))
marriage_dict = pickle.load(open("pickles/marriage_dictionary.p", "rb"))
overall_dict = pickle.load(open("pickles/overall_dictionary.p", "rb"))
```

```python
df = pd.read_csv('data/cleaned_snowball.csv', sep=',', quotechar='"', index_col=0, header=0)
```

### Setup the custom part of the tokenizer and apply it to the tweet

```python
keep = set(['!','?'])
stop = set(stopwords.words('english'))
remove = set([x for x in list(string.punctuation) if x not in keep])
stop.update(remove)
stop.update(['',' ','  '])
```

### Get the topics

```python
df['climate_topics'] = df[df.topic == 'C'].text_processed.apply(lambda x: climate_model.get_document_topics(climate_model.id2word.doc2bow([d for d in x.split(" ") if d not in stop]), per_word_topics=False))
df['guns_topics'] = df[df.topic == 'G'].text_processed.apply(lambda x: guns_model.get_document_topics(guns_model.id2word.doc2bow([d for d in x.split(" ") if xd not in stop]), per_word_topics=False))
df['marriage_topics'] = df[df.topic == 'M'].text_processed.apply(lambda x: marriage_model.get_document_topics(marriage_model.id2word.doc2bow([d for d in x.split(" ") if d not in stop]), per_word_topics=False))
df['overall_topics'] = df.text_processed.apply(lambda x: overall_model.get_document_topics(overall_model.id2word.doc2bow([d for d in x.split(" ") if d not in stop]), per_word_topics=False))
```
